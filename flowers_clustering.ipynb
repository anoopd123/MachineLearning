# Clustering

```python
import pandas as pd
```

```python
df = pd.read_csv("flowers.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
clusters = 2
```

```python
model = KMeans(n_clusters = clusters)
```

```python
model.fit(df)
```

```python
model.predict(df)
```

```python
df['cluster'] = model.predict(df)
```

```python
df.head()
```

```python
df.cluster.value_counts()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df, hue = 'cluster')  # hue = 'cluster' is used to change the color of the cluster based on the provided number of clusters
```
